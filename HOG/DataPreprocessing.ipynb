```python
# Import the functions to calculate feature descriptors
from skimage.feature import local_binary_pattern
from skimage.feature import hog
from skimage.io import imread, imsave
from sklearn.externals import joblib
from skimage.transform import resize
from numpy import random
# To read file names
import glob
import os
# To display the plot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pos_im_path = '../TrainingData'
pos_dest_path = '../TrainingData/positive'
im_path ='../TrainingData/2018-04-03_16-13-25.jpg'
im = imread(im_path)#, as_grey=False)
im2 = resize(im,(64,64))
plt.imshow(im2)#, cmap='gray') 

imsave(fname = pos_dest_path+'/1.jpg', arr = im2, quality = 100)
```

```python
i = 1
for im_path in glob.glob(os.path.join(pos_im_path, "*.png")):     
    im = imread(im_path)#, as_grey=True)
    im2 = resize(im,(64,64))
    filename = pos_dest_path+'/'+str(i)+'.jpg'
    imsave(fname = filename, arr = im2, quality = 100)
    i +=1
```

```python
IMG_WIDTH = 64
IMG_LENGTH = 64

pos_im_path = '../session3'
neg_dest_path = '../TrainingData/negative'
im_path ='../session3/DJI_0038 40.jpg'
im = imread(im_path)#, as_grey=False)
# im2 = resize(im,(64,64))
plt.imshow(im)#, cmap='gray') 
width,length = im.shape[0:2]
n = 60
for i in range(n,2*n):
    w_start = random.randint(width-IMG_WIDTH)
    l_start = random.randint(length-IMG_LENGTH)
    im_crop = im[w_start:(w_start+IMG_WIDTH),l_start:(l_start+IMG_LENGTH),:]
    imsave(fname = neg_dest_path+'/'+str(i)+'.jpg', arr = im_crop, quality = 100)
```
